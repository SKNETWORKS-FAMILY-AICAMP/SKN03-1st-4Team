```python
import pandas as pd # 데이터 추출, 재가공, 정렬을 위한 모듈

df = pd.read_excel("./car_info/전국 자동차 등록 현황/승용/개인승용.xlsx") # 엑셀 파일 불러오기

test_df = df.iloc[range(0,54),range(0,19)]  # [1, 19, 36, 53, 87]행, 3~20열 추출
test_df
```
